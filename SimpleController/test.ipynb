```python
import numpy as np

def points_on_circle(num_points, radius):
    points = []
    for i in range(num_points):
        angle = 2 * np.pi * i / num_points
        x = radius * np.cos(angle)
        y = radius * np.sin(angle)
        points.append((x, y))
    return points

# Number of points and radius
num_points = 5
radius = 2

# Calculate points
points = points_on_circle(num_points, radius)

# Print coordinates of points
for i, point in enumerate(points, start=1):
    print(f"Point {i}: {np.around(point,2)}")
```
